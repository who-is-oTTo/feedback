```python
%pip install numpy

def homework(message, answer = ''):
    print(f'\n‚ùì {message}\nüí° {answer}\n')
```

```python
import numpy as np
temperature = [20.1, 20.1, 20.2, 20.4]
print(temperature)

t1 = temperature * 5

print(f"–ü—Ä–∏ –º–Ω–æ–∂–µ–Ω–Ω—ñ –Ω–∞ 5 –º–∞—î–º–æ: {t1}")
homework("–ß–æ–º—É –º–∞—î–º–æ —Ç–∞–∫–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç?", "–í—Å–µ –ø—Ä–∞–≤–∏–ª—å–Ω–æ, –º–∞—Å–∏–≤ –ø–æ–º–Ω–æ–∂–∏–≤—Å—è –Ω–∞ 5, —Ç–æ–±—Ç–æ –ø–æ–≤—Ç–æ—Ä–∏–≤—Å—è 5 —Ä–∞–∑—ñ–≤")

t1 = [ x*5 for x in temperature] 
print(f"–¢–µ–ø–µ—Ä –æ—Ç—Ä–∏–º–∞—î–º–æ: {t1}") 
homework("–ß–æ–º—É —Ç–µ–ø–µ—Ä –º–Ω–æ–∂–µ–Ω–Ω—è —É—Å–ø—ñ—à–Ω–µ?", "–ë–æ —Ç–µ–ø–µ—Ä –≥–µ–Ω–µ—Ä—É—î—Ç—å—Å—è –º–∞—Å–∏–≤ –≤ —è–∫–æ–º—É –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç –ø–æ–º–Ω–æ–∂–µ–Ω–∏–π –Ω–∞ 5")

t2 = np.array(temperature)
print("–ú–∞—Å–∏–≤ –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –º–µ—Ç–æ–¥—ñ–≤ NumPy", t2)

t3 = t2 * 5
print("–ü—Ä–∏ –º–Ω–æ–∂–µ–Ω–Ω—ñ –Ω–∞ 5 –æ—Ç—Ä–∏–º—É—î–º–æ:", t3)

homework('–¶—ñ–∫–∞–≤–æ )', '–°—Ö–æ–∂–µ –ù–∞–º–ü–∞–π –¥—É–∂–µ —Ä–æ–∑—É–º–Ω–∏–π, —è–∫—â–æ —Ä–æ–∑—É–º—ñ—î —â–æ —Ç—Ä–µ–±–∞ —Ä–æ–±–∏—Ç–∏ )')
```

### –ú–µ—Ç–æ–¥ arange - –≥–µ–Ω–µ—Ä—É—î –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å —á–∏—Å–µ–ª –∑ –ø–µ–≤–Ω–∏–º –∫—Ä–æ–∫–æ–º —É –∑–∞–¥–∞–Ω–æ–º—É —ñ–Ω—Ç–µ—Ä–≤–∞–ª—ñ

```python
start = 5
size_of_vec = 100
x = range(size_of_vec)
a = np.arange(size_of_vec)
print(f"–ú–∞—Å–∏–≤ –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–∏–π –ø—Ä–æ—Å—Ç–∏–º range: {x[0:5]} —Ç–∞ –º–µ—Ç–æ–¥–æ–º arange: {a[0:5]}")
homework("–í–∏–≤–µ–¥—ñ—Ç—å –≤–µ—Å—å –¥—ñ–∞–ø–∞–∑–æ–Ω —á–∏—Å–µ–ª.", a)

a = np.arange(start, size_of_vec)
print(f"–ú–æ–∂–µ–º–æ –∑–∞–¥–∞–≤–∞—Ç–∏ –ø–æ—á–∞—Ç–æ–∫ –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó –≤–µ–∫—Ç–æ—Ä–∞, –ø–µ—Ä—à–∏–π –µ–ª–µ–º–µ–Ω—Ç: {a[0]}")
homework("–í–∏–≤–µ–¥—ñ—Ç—å –æ—Å—Ç–∞–Ω–Ω—ñ–π –¥—ñ–∞–ø–∞–∑–æ–Ω –≤ –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–æ–º—É –º–∞—Å–∏–≤—ñ.", a[-1])

x = range(start, size_of_vec)
print("–¢–µ —Å–∞–º–µ –∞–ª–µ –∑ –≤–±—É–¥–æ–≤–∞–Ω–æ—é —Ñ—É–Ω–∫—Ü—ñ—î—é range:", x)
print("–í–ª–∞—Å–Ω–µ –≤–∏–≤—ñ–¥ –∑–Ω–∞—á–µ–Ω—å:", [0], )
homework("–í–∏–≤–µ–¥—ñ—Ç—å —è–∫–µ—Å—å —á–∏—Å–ª–æ –∑ —Å–µ—Ä–µ–¥–∏–Ω–∏ –¥—ñ–∞–ø–∞–∑–æ–Ω—É.", x[13])

# –¥—Ä–æ–±–æ–≤—ñ —á–∏—Å–ª–∞
x = np.arange(10.4)
print("–ë—É–¥–µ–º –º–∞—Ç–∏:", x)

# –∫—Ä–æ–∫
x = np.arange(0.5, 4, 0.5)
print("–ó –∫—Ä–æ–∫–æ–º 0.5:", x)

# –ó–∞–æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è
x = np.arange(0.04, 1.84, 0.08)
print("–ó –∫—Ä–æ–∫–æ–º 0.08:", x, "\n –ó–µ—Ä–Ω—ñ—Ç—å —É–∞–≥—É –Ω–∞ –∑–∞–æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è")

vec = np.arange( 0.00001, 0.0001, 0.00001)
homework("–ó–≥–µ–Ω–µ—Ä—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–µ –∫—Ä–æ–∫ –±—É–¥–µ –º–∞—Ç–∏ —á–æ—Ç–∏—Ä–∏ –Ω—É–ª—ñ –ø—ñ—Å–ª—è –∫–æ–º–∏, —è–∫–∏–º –±—É–¥–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç?", vec)

y = x * 2.5
print("–ü—ñ—Å–ª—è –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–Ω—è:", y)

res = (y ** 2) / 0.4
homework("–ü—ñ–¥–Ω–µ—Å—ñ—Ç—å –æ—Ç—Ä–∏–º–∞–Ω–µ —á–∏—Å–ª–æ –¥–æ —Å—Ç–µ–ø–µ–Ω—è —Ç–∞ –ø–æ–¥—ñ–ª—ñ—Ç—å –Ω–∞ 0.4 - –Ø–∫–∏–º –±—É–¥–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç?", res)
```

### –ú–µ—Ç–æ–¥ linspace - –≥–µ–Ω–µ—Ä—É—î –∑–∞–¥–∞–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—é —á–∏—Å–µ–ª —É –ø–µ–≤–Ω–æ–º—É –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ, 50 —á–∏—Å–µ–ª –∑–∞ –∑–∞–º–æ–≤—á—É–≤–∞–Ω–Ω—è–º

```python
a = np.linspace(start, size_of_vec)
print(f"–ó–≥–µ–Ω–µ—Ä—É–≤–∞–ª–∏ {len(a)} —á–∏—Å–µ–ª –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ {a[0]} –¥–æ {a[-1]}")
#print(f"–í–ª–∞—Å–Ω–µ —á–∏—Å–ª–∞: {a}")

a2 = list()
for i in range(0, len(a), 2):
    a2.append(a[i])

homework("–í–∏–≤–µ–¥—ñ—Ç—å –∫–æ–∂–Ω–µ –¥—Ä—É–≥–µ —á–∏—Å–ª–æ –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–æ–≥–æ –¥—ñ–∞–ø–∞–∑–æ–Ω—É. –¢—É—Ç –±—É–¥–µ–µ –ø–æ—Ç—Ä—ñ–±–µ–Ω —Ü–∏–∫–ª :)", a2)

print("–ú–æ–∂–Ω–∞ –∑–º–µ–Ω—à–∏—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª –¥–æ 5:")
a = np.linspace(start, size_of_vec, 5)
print(a)
homework("–ó–±—ñ–ª—å—à—ñ—Ç—å –∫—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª –¥–æ 100. –í–∏–≤–µ–¥—ñ—Ç—å –ø–µ—Ä—à–∏—Ö —Ç–∞ –æ—Å—Ç–∞–Ω–Ω—ñ—Ö 10.")

print('–ö—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª 100:')
a = np.linspace(start, size_of_vec, 100)
print(a)
print('–ü–µ—Ä—à—ñ 10:\n', a[0:10])
print('–û—Å—Ç–∞–Ω–Ω—ñ 10:\n', a[-10:])
print()

print("""–Ø–∫ –±–∞—á–∏–º–æ –ø–µ—Ä—à–µ —Ç–∞ –æ—Å—Ç–∞–Ω–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –∑–∞–≤–∂–¥–∏ –≤—Ö–æ–¥—è—Ç—å —É –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–∏–π –¥—ñ–∞–ø–∞–∑–æ–Ω. 
–í—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –ø—Ä–∏ —Ü—å–æ–º—É –≤–∏—Ä–∞—Ö–æ–≤—É—î—Ç—å—Å—è –∫—Ä–æ–∫.
–¶–µ –º–æ–∂–Ω–∞ –∑–º—ñ–Ω–∏—Ç–∏, –≤–∏–∫–ª—é—á–∏–≤—à–∏ –æ—Å—Ç–∞–Ω–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –¥—ñ–∞–ø–∞–∑–æ–Ω—É.
""")

print(np.linspace(start, size_of_vec, 5, endpoint=False))
homework("–°–ø—Ä–æ–±—É–π—Ç–µ —Å–∞–º—ñ –ø–æ—Ä–∞—Ö—É–∞–≤—Ç–∏ –∫—Ä–æ–∫.", "–ó–¥–∞—î—Ç—å—Å—è –∫—Ä–æ–∫ 19. –¶–µ —á–æ–º—É —Ç–∞–∫ ü§î ") 

_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
print(_)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)
print(_)

print(f"–î–ª—è –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è, –º–∏ –º–æ–∂–µ–º–æ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –∫—Ä–æ–∫ –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ, —É –Ω–∞—à–æ–º—É –≤–∏–ø–∞–¥–∫—É —Ü–µ {sp_1} —Ç–∞ {sp_2}")
homework("–ß–æ–º—É –º–∞—î–º–æ —Ç–∞–∫–∏–π –∫—Ä–æ–∫?", "–¢—Ä–æ—à–∫–∏ –¥–∏–≤–Ω–∞ —à—Ç—É–∫–∞ ü§î\n   –í –¥—Ä—É–≥–æ–º—É –≤–∏–ø–∞–¥–∫—É –≤–∏—Ö–æ–¥–∏—Ç—å 95 –µ–ª–µ–º–Ω—Ç—ñ–≤, –±–æ –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ 5 —ñ –π–¥–µ –¥–æ 100. 95 / 5 = 19\n   –ê –æ—Ç –∑ –ø–µ—Ä—à–∏–º –Ω–µ –¥—É–∂–µ —Ä–æ–∑—ñ–±—Ä–∞–≤—Å—è.")
```

### –¢–∏–ø–∏ –º–∞—Å–∏–≤—ñ–≤/–≤–µ–∫—Ç–æ—Ä—ñ–≤

```python
examples = {
    '–°–∫–∞–ª—è—Ä': np.array(1), 
    '–í–µ–∫—Ç–æ—Ä –†—è–¥–æ–∫': np.array([1, 2, 3]), 
    '–í–µ–∫—Ç–æ—Ä –°—Ç–æ–≤–ø–µ—Ü—å': np.array([[1], [2], [3]]), 
    '–ú–∞—Ç—Ä–∏—Ü—è': np.array([[1, 2],[3, 4], [5, 6]]),
    '–ë–∞–≥–∞—Ç–æ–≤–∏–º—ñ—Ä–Ω–∞ –ú–∞—Ç—Ä–∏—Ü—è': np.array([ [[111, 112], [121, 122]],
                                        [[211, 212], [221, 222]],
                                        [[311, 312], [321, 322]] ]),
    '–î—É—É–∂–µ–µ –ë–∞–≥–∞—Ç–æ–≤–∏–º—ñ—Ä–Ω–∞ –ú–∞—Ç—Ä–∏—Ü—è –∑ 1 –µ–ª–µ–º–µ–Ω—Ç–æ–º )': np.array([[[[[[[[[[1]]]]]]]]]])
    }

for key in examples:
    print(f"–†–æ–∑–≥–ª—è–¥–∞—î–º–æ –ø—Ä–∏–∫–ª–∞–¥ –∑ {key}. \n–ó–Ω–∞—á–µ–Ω–Ω—è x –±—É–¥–µ: \n", examples[key])
    print(f"–¢–∏–ø {key} x: ", type(examples[key]))
    print(f"–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å/–≤–∏–º—ñ—Ä {key} x:", np.ndim(examples[key]))
    print(f"–§–æ—Ä–º–∞ {key} x: ", examples[key].shape)
    print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Ö: {np.size(examples[key])}")
    print("-=-"*10)

homework("""–Ø–∫ –í–∏ –¥—É–º–∞—î—Ç–µ —á–æ–º—É –º–∞—î–º–æ —Ç–∞–∫–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó np.ndim (–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å/–≤–∏–º—ñ—Ä/dimension).
–ß–æ–º—É –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è –≤–µ–∫—Ç–æ—Ä—ñ–≤ —Ç–∞ –º–∞—Ç—Ä–∏—Ü—å –æ–¥–Ω–∞–∫–æ–≤–µ —Ç–∞ –Ω–µ —Å–ø—ñ–≤–ø–¥–∞—î –∑ –∑–Ω–∞—á–µ–Ω–Ω—è–º shape?
–ó–∞–¥–∞–π—Ç–µ –≤–ª–∞—Å–Ω—ñ –≤–µ–∫—Ç–æ—Ä–∏ –∑ —Ä—ñ–∑–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏.""", "np.ndim –ø–æ–≤–µ—Ä—Ç–∞—î –≥–ª–∏–±–∏–Ω—É –º–∞—Å–∏–≤—É, —Ç–æ–±—Ç–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏–º—ñ—Ä—ñ–≤, –∞ –º–µ—Ç–æ–¥ shape –ø–æ–≤–µ—Ä—Ç–∞—î —Ñ–æ—Ä–º—É ü§î")
```

### –ë–∞–≥–∞—Ç–æ–≤–∏–º—ñ—Ä–Ω—ñ –º–∞—Å–∏–≤–∏, —Ä–æ–∑—É–º—ñ–Ω–Ω—è –≤–∏–º—ñ—Ä—ñ–≤

```python
examples = {
    '5 —Ä—è–¥–∫—ñ–≤': np.array([[1], [2], [3], [4], [5]]), 
    '5 —Å—Ç–æ–≤–ø—Ü—ñ–≤': np.array([[1, 2, 3, 4, 5]]),
    '5 –µ–ª–µ–º–µ–Ω—Ç—ñ–≤': np.array([[[1, 2, 3, 4, 5]]])
    }

for a in examples:
    print(f"–î–ª—è {a} –º–∞—î–º–æ shape: {examples[a].shape} \n—Ç–∞ –≤–∏–≥–ª—è–¥: \n{examples[a]}\n")

homework(f"–°—Ç–≤–æ—Ä—ñ—Ç—å –≤–ª–∞—Å–Ω—ñ –º–∞—Å–∏–≤–∏ –∑ —Ä—ñ–∑–Ω–æ—é –∫—ñ–ª—å–∫—ñ—Å—Ç—é –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Ç–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ –º–µ—Ç–æ–¥ shape –¥–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è —ó—Ö –≤–∏–º—ñ—Ä—ñ–≤.")

shps = {
    '–ö–≤–∞–¥—Ä–∞—Ç' : np.array([[1, 2],[3, 4]]),
    '–ö—É–±' : np.array([[[1, 2],[3, 4]], [[5, 6], [7, 8]]])
}

for s in shps:
    print(f"{s}\n{shps[s].shape}\n{shps[s]}\n")
```

```python
F = np.array([1, 1, 2, 3, 5, 8, 13, 21])
print(f"""–ú–∞—Å–∏–≤: {F}
–í–∏–º—ñ—Ä–∏: {F.shape}
–ü–∞–º—è—Ç–∞—î–º–æ, —â–æ –Ω–æ–º–µ—Ä–∞—Ü—ñ—è –º–∞—Å–∏–≤—ñ–≤ –≤—ñ–¥–±—É–≤–∞—î—Ç—å—Å—è –∑ 0.
–¢–æ–º—É –ø–µ—Ä—à–∏–π –µ–ª–µ–º–µ–Ω—Ç –º–∞—Å–∏–≤—É:\n {F[0]}
–î—Ä—É–≥–∏–π:\n {F[1]}
–û—Å—Ç–∞–Ω–Ω—ñ–π, {len(F)}: \n {F[7]}""")

print(f"""–ú–æ–∂–µ–º–æ —Å–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏—Å—å –≤—ñ–¥—î–º–Ω–∏–º–∏ —ñ–Ω–¥–µ–∫—Å–∞–º–∏ —è–∫—â–æ —Ö–æ—á–µ–º–æ –¥–æ—Å—Ç—É–ø–∏—Ç–∏—Å—å –¥–æ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∑ –∫—ñ–Ω—Ü—è –º–∞—Å–∏–≤—É.
–ê–ª–µ –Ω–æ–º–µ—Ä–∞—Ü—ñ—è –≤–∂–µ –∑ 1.
–¢–æ–º—É –æ—Å—Ç–∞–Ω–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç:\n {F[-1]}
–ü–µ—Ä–µ–¥–æ—Å—Ç–∞–Ω–Ω—ñ–π:\n {F[-2]}""")

homework("–ü–æ–ø—Ä–∞–∫—Ç–∏–∫—É–π—Ç–µ—Å—å –¥–æ—Å—Ç—É–ø–∞—Ç–∏—Å—å –¥–æ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, –≤–∏–≤–µ–¥—ñ—Ç—å –¥–µ–∫—ñ–ª—å–∫–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Ä—ñ–∑–Ω–∏–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏.")

array = [ i for i in range(10) ]
print(f'{array = }')
print(array[0]) # –≤–∏–≤–æ–¥–∏—Ç—å 1 –µ–ª–µ–º
print(array[:0]) # –≤–∏–≤–æ–¥–∏—Ç—å 0 –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ )
print(array[6:]) # –≤–∏–≤–æ–¥–∏—Ç—å –µ–ª–µ–º–µ–Ω—Ç–∏ –∑ 6
print(array[3:6]) # –≤–∏–≤–æ–¥–∏—Ç—å –∑ 3 –ø–æ 6 –µ–ª–µ–º
print(array[::]) # –≤–∏–≤–æ–¥–∏—Ç—å –≤—Å—ñ –µ–ª–µ–º–µ–Ω—Ç–∏
print(array[0:10:2]) # –æ–≥–æ, –≤–æ–Ω–æ –≤–∏–≤–æ–¥–∏—Ç—å –∫–æ–∂–µ–Ω 2 –µ–ª–µ–º–µ–Ω—Ç –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 0 —ñ –¥–æ 10

```

```python
A = np.arange(50)
all_even = (A%2==0) # –ø–∞—Ä–Ω—ñ
complex = (A<30) & (A/5<2) 

print(f"""–°–ø–æ—á–∞—Ç–∫—É –ø–æ—Ç—Ä—ñ–±–Ω–æ —Å—Ç–≤–æ—Ä–∏—Ç–∏ —É–º–æ–≤—É –∑–∞ —è–∫–æ—é –∑–¥—ñ–π—Å–Ω—é–≤–∞—Ç–∏ –≤–∏–±—ñ—Ä–∫—É.
–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–æ–≤–∏ –±—É–¥–µ –∞–±–æ {all_even[0]} –¥–ª—è {A[0]} –∞–±–æ {all_even[1]} –¥–ª—è {A[1]}.
–ü—ñ—Å–ª—è –ø—ñ–¥—Å—Ç–∞–≤–ª—è–Ω–Ω—è —É–º–æ–≤–∏ –ø–æ—à—É–∫—É –ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª –≤ –º–∞—Å–∏–≤ –æ—Ç—Ä–∏–º–∞—î–º–æ:
{A[all_even]}
–ê–±–æ –∑–∞—Å—Ç–æ—Å—É–≤–∞–≤—Ç–∏ –±—ñ–ª—å—à –∫–æ–º–ø–ª–µ–∫—Å–Ω—É —É–º–æ–≤—É –∑ –¥–≤–æ—Ö –ø–æ—Ä—ñ–Ω—è–Ω—å:
{A[complex]}""")

homework("–í–∏–≤–µ–¥—ñ—Ç—å –≤—Å—ñ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑ –¥–∞–Ω–æ–≥–æ –º–∞—Å–∏–≤—É.")

all_not_even = (A%2==1)
print(f'–í—Å—ñ –Ω–µ –ø–∞—Ä–Ω—ñ:\n {A[all_not_even]}')
```
